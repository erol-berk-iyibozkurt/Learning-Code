### Downloading the stock data of AAPL (Apple Inc.)

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker = 'AAPL'
# Download the historical data from the ticker
data = yf.download(ticker,start="2010-01-01",end="2024-09-30")
# Display the first few rows of the data
print(data.head())
# Save the to a CSV file
data.to_csv(f'{ticker}_historical_data.csv')
```

```python
import yfinance as yf
import pandas as pd
import csv 


df = pd.read_csv("companiesmarketcap.com - Largest tech companies by market cap.csv")
selected_stock_list = list(df.iloc[:49]["Symbol"])

def download_stock_data():
    for i in selected_stock_list:
    # Download the historical data from the ticker
        data = yf.download(i,start="2010-01-01",end="2024-09-30")
    # Save the to a CSV file
        data.to_csv(f'{i}_data.csv')

download_stock_data()
```
