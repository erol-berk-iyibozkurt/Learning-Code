```python
import yfinance as yf
import pandas as pd

# Döviz kuru sembolü (KRW → USD)
currency_pair = 'KRWUSD=X'

# Döviz kuru verilerini indirme ve kaydetme fonksiyonu
def download_currency_data(currency_pair, start_date, end_date):
    print(f"{currency_pair} döviz kuru verileri indiriliyor...")
    
    try:
        # Döviz kuru verilerini indir
        data = yf.download(currency_pair, start=start_date, end=end_date)
        
        if not data.empty:
            # Veriyi CSV olarak kaydet
            data.to_csv(f'{currency_pair}_data.csv')
            print(f"{currency_pair} döviz kuru verileri başarıyla kaydedildi: {currency_pair}_data.csv")
        else:
            print("Döviz kuru verisi boş döndü. Tarih aralığını kontrol edin.")
    
    except Exception as e:
        print(f"{currency_pair} için veri indirme sırasında hata oluştu: {e}")

# Fonksiyonu çağır
download_currency_data(currency_pair='KRWUSD=X', start_date='2010-01-01', end_date='2024-09-30')
```

```python
import yfinance as yf
import pandas as pd

# CSV dosyasını yükle
df = pd.read_csv("/Users/erolberkiyibozkurt/Documents/GitHub/Python/Projects/The Influence of AI Boom on Tech Stocks/Background Data/companiesmarketcap.com - Largest tech companies by market cap.csv")

# İlk 100 şirketin sembollerini al
selected_stock_list = list(df.iloc[:100]["Symbol"])

# Döviz türlerini saklamak için boş bir liste oluştur
currency_info = []

# Döviz türlerini belirleme fonksiyonu
def get_currency_for_stocks():
    for symbol in selected_stock_list:
        print(f"{symbol} döviz türü kontrol ediliyor...")
        try:
            stock = yf.Ticker(symbol)
            stock_info = stock.info
            currency = stock_info.get('currency', 'Bilinmiyor')  # Döviz türü yoksa 'Bilinmiyor'
            
            # Şirket ve döviz bilgilerini listeye ekle
            currency_info.append({'Symbol': symbol, 'Currency': currency})
        
        except Exception as e:
            print(f"{symbol} için döviz türü alınamadı: {e}")
            currency_info.append({'Symbol': symbol, 'Currency': 'Hata'})
    
    # Veriyi DataFrame'e dönüştür
    currency_df = pd.DataFrame(currency_info)
    currency_df.to_csv('stock_currency_info.csv', index=False)
    print("Tüm döviz bilgileri 'stock_currency_info.csv' dosyasına kaydedildi.")
    return currency_df

# Fonksiyonu çalıştır ve sonucu göster
currency_df = get_currency_for_stocks()
print(currency_df)
```

```python
import yfinance as yf
import pandas as pd

# Load the CSV file
file_path = "/Users/erolberkiyibozkurt/Documents/GitHub/Python/Projects/The Influence of AI Boom on Tech Stocks/Background Data/companiesmarketcap.com - Largest tech companies by market cap.csv"
df = pd.read_csv(file_path)

# Get the symbols of the top 100 companies
selected_stock_list = list(df.iloc[:100]["Symbol"])

# Create an empty list to store stock information
stock_info_list = []

# Function to retrieve stock exchange and currency information
def get_exchange_and_currency():
    for symbol in selected_stock_list:
        print(f"Checking exchange and currency information for {symbol}...")
        try:
            stock = yf.Ticker(symbol)
            stock_info = stock.info
            
            # Retrieve exchange and currency information
            exchange = stock_info.get('exchange', 'Unknown')
            currency = stock_info.get('currency', 'Unknown')
            country = stock_info.get('country', 'Unknown')
            
            # Special case: If listed on NASDAQ or NYSE, set currency to USD
            if exchange in ['NASDAQ', 'NYSE']:
                currency = 'USD'
            
            # Add information to the list
            stock_info_list.append({
                'Symbol': symbol,
                'Exchange': exchange,
                'Currency': currency,
                'Country': country
            })
        
        except Exception as e:
            print(f"Failed to retrieve data for {symbol}: {e}")
            stock_info_list.append({
                'Symbol': symbol,
                'Exchange': 'Error',
                'Currency': 'Error',
                'Country': 'Error'
            })
    
    # Convert the list into a DataFrame
    stock_info_df = pd.DataFrame(stock_info_list)
    output_path = "/Users/erolberkiyibozkurt/Documents/GitHub/Python/Projects/The Influence of AI Boom on Tech Stocks/Background Data/stock_exchange_currency_info.csv"
    stock_info_df.to_csv(output_path, index=False)
    print(f"All exchange and currency information has been saved to '{output_path}'.")
    return stock_info_df

# Execute the function and display the result
if __name__ == "__main__":
    stock_info_df = get_exchange_and_currency()
    print("Final output preview:")
    print(stock_info_df.head())
```

```python
import yfinance as yf
import pandas as pd

# Define the currencies and their forex pairs
non_usd_currencies = ['KRW', 'HKD', 'EUR', 'JPY', 'ILS', 'TWD', 'CAD', 'THB']
forex_pairs = [f"{currency}USD=X" for currency in non_usd_currencies]

# Download forex data for each pair
forex_data = {}
for pair in forex_pairs:
    print(f"Downloading forex data for {pair}...")
    try:
        # Explicitly call yf.download without overwriting it accidentally
        forex_response = yf.download(pair, start="2024-01-01", end="2024-09-30")
        if not forex_response.empty:
            forex_data[pair] = forex_response['Close']
        else:
            print(f"No data available for {pair}. Skipping...")
    except Exception as e:
        print(f"Failed to download data for {pair}: {e}")

# Combine valid forex data into a DataFrame
if forex_data:
    forex_df = pd.concat(forex_data, axis=1)
    forex_output_path = "/Users/erolberkiyibozkurt/Documents/GitHub/Python/Projects/The Influence of AI Boom on Tech Stocks/Background Data/forex_rates.csv"
    forex_df.to_csv(forex_output_path)
    print(f"Forex rates saved to {forex_output_path}")
else:
    print("No valid forex data was downloaded. Please check the currency pairs.")
```
