```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi okuma
column_names = ['Date', 'Price', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
df = pd.read_csv('AAPL_historical_data.csv', skiprows=2, names=column_names, parse_dates=['Date'], index_col='Date')

# Grafik oluşturma
plt.figure(figsize=(12, 6))
df['Adj Close'].plot()

# Versiyon tarihlerine göre taralı alan ekleme
plt.axvspan(pd.Timestamp('2023-03-14'), pd.Timestamp('2023-04-01'), color='orange', alpha=0.3, label='GPT-4 Release')
plt.axvspan(pd.Timestamp('2023-06-14'), pd.Timestamp('2023-07-01'), color='blue', alpha=0.3, label='GPT-4 Turbo Release')
plt.axvspan(pd.Timestamp('2024-03-14'), pd.Timestamp('2024-04-01'), color='green', alpha=0.3, label='Next Version')

# Grafik detayları
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('Apple Stock Price Timeseries with ChatGPT Releases')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Veriyi okuma
column_names = ['Date', 'Price', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
df = pd.read_csv('AAPL_historical_data.csv', skiprows=2, names=column_names, parse_dates=['Date'], index_col='Date')

# Grafik oluşturma
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(df['Adj Close'], color='purple', linewidth=1.5, label='Adjusted Close Price')

# Versiyon tarihlerine göre taralı alan ekleme
# Tarihleri matplotlib date numaralarına dönüştürme
ax.axvspan(mdates.date2num(datetime(2023, 3, 14)), mdates.date2num(datetime(2023, 4, 1)), 
           color='orange', alpha=0.2, label='GPT-4 Release')
ax.axvspan(mdates.date2num(datetime(2023, 6, 14)), mdates.date2num(datetime(2023, 7, 1)), 
           color='blue', alpha=0.2, label='GPT-4 Turbo Release')
ax.axvspan(mdates.date2num(datetime(2024, 3, 14)), mdates.date2num(datetime(2024, 4, 1)), 
           color='green', alpha=0.2, label='Next Version')

# X ekseni ayarları
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Adjusted Close Price (USD)', fontsize=12)
ax.set_title('Apple Stock Price with ChatGPT Release Dates', fontsize=14, fontweight='bold')

# Tarih formatını ve aralığını ayarlayın
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.xticks(rotation=45, ha='right')

# Eksen çizgileri ve ızgara
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
ax.tick_params(axis='both', which='major', labelsize=10)

# Efsane ayarları
plt.legend(loc='upper left', fontsize=10, frameon=True, shadow=True)

# Grafik gösterme
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Veriyi okuma
column_names = ['Date', 'Price', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
df = pd.read_csv('AAPL_historical_data.csv', skiprows=2, names=column_names, parse_dates=['Date'], index_col='Date')

# Grafik oluşturma
plt.figure(figsize=(14, 7))
plt.plot(df['Adj Close'], color='purple', linewidth=1.5, label='Adjusted Close Price')

# Önemli GPT sürüm tarihleri için taralı alanlar
plt.axvspan(datetime(2023, 3, 14), datetime(2023, 4, 1), color='#FF8C00', alpha=0.4, label='GPT-4 Release')
plt.axvspan(datetime(2023, 6, 14), datetime(2023, 7, 1), color='#1E90FF', alpha=0.4, label='GPT-4 Turbo Release')
plt.axvspan(datetime(2024, 3, 14), datetime(2024, 4, 1), color='#00CD66', alpha=0.4, label='Next GPT Version')

# X ekseni ayarları
plt.xlabel('Date', fontsize=12)
plt.ylabel('Adjusted Close Price (USD)', fontsize=12)
plt.title('Apple Stock Price with Major GPT Model Release Dates', fontsize=14, fontweight='bold')

# Tarih formatını ve aralığını ayarlayın
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.xticks(rotation=45, ha='right')

# Eksen çizgileri ve ızgara
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tick_params(axis='both', which='major', labelsize=10)

# Efsane ayarları
plt.legend(loc='upper left', fontsize=10, frameon=True, shadow=True)

# Grafik gösterme
plt.tight_layout()
plt.show()
```
